```python
def safe_float_conversion(value, default=None):
    """ Attempt to convert a value to float; use default if conversion fails. """
    try:
        return float(value)
    except ValueError:
        return default

# Find a default value for 'person_emp_length' - using median as an example
employment_lengths = [safe_float_conversion(row['person_emp_length']) for row in data if row['person_emp_length'] not in ('', None)]
median_emp_length = sorted(employment_lengths)[len(employment_lengths) // 2]

# Convert and clean 'person_emp_length'
for row in data:
    if row['person_emp_length'] not in ('', None):
        row['person_emp_length'] = safe_float_conversion(row['person_emp_length'], median_emp_length)
    else:
        row['person_emp_length'] = median_emp_length

# Convert and clean 'loan_int_rate'
for row in data:
    if row['loan_int_rate'] not in ('', None):
        row['loan_int_rate'] = safe_float_conversion(row['loan_int_rate'])
    else:
        # If you decide to handle missing rates differently, adjust here
        row['loan_int_rate'] = None  # or some other default value, such as the median of existing rates
```

## hypothesis testing (chi_square statistics)

### age vs loan status

```python
# Extract relevant data for age and loan status
ages = [int(row['person_age']) for row in data]
loan_statuses = [int(row['loan_status']) for row in data]

# Group ages and corresponding loan statuses
age_loan_statuses = {}
for age, loan_status in zip(ages, loan_statuses):
    if age in age_loan_statuses:
        age_loan_statuses[age]['total'] += 1
        age_loan_statuses[age]['defaults'] += loan_status
    else:
        age_loan_statuses[age] = {'total': 1, 'defaults': loan_status}

# Calculate proportion of defaulted loans for each age group
age_default_proportion = {age: age_info['defaults'] / age_info['total'] for age, age_info in age_loan_statuses.items() if age_info['total'] != 0}

# Sort age groups based on age
sorted_ages = sorted(age_default_proportion.keys())
sorted_proportions = [age_default_proportion[age] for age in sorted_ages]
sorted_non_default_proportions = [1 - proportion for proportion in sorted_proportions]

# Create stacked bar chart
plt.figure(figsize=(10, 6))
plt.bar(sorted_ages, sorted_non_default_proportions, color='lightgreen', label='Non-Defaulted')
plt.bar(sorted_ages, sorted_proportions, bottom=sorted_non_default_proportions, color='lightcoral', label='Defaulted')
plt.xlabel('Age')
plt.ylabel('Proportion of Loans')
plt.title('Stacked Bar Chart: Age vs Loan Status')
plt.legend()
plt.grid(axis='y')
plt.show()
```

**hypothesis testing :**

Null Hypothesis (H0): There is no association between age groups and loan defaults.

Alternative Hypothesis (H1): There is an association between age groups and loan defaults.

We assume, significance threshold is 0.05. If p value is less than significance, then we'll reject the null hypothesis.

```python
from scipy.stats import chi2_contingency

# Categorize individuals into age groups
# Assume age groups: < 30, 30-40, 40-50, 50-60, >= 60
age_groups = {'< 30': [], '30-40': [], '40-50': [], '50-60': [], '>= 60': []}
for row in data:
    age = int(row['person_age'])
    if age < 30:
        age_group = '< 30'
    elif age < 40:
        age_group = '30-40'
    elif age < 50:
        age_group = '40-50'
    elif age < 60:
        age_group = '50-60'
    else:
        age_group = '>= 60'
    age_groups[age_group].append(int(row['loan_status']))

# Convert dictionary values to lists
observed = [[sum(age_groups[group]), len(age_groups[group]) - sum(age_groups[group])] for group in age_groups.keys()]

# Perform chi-square test
chi2, p, _, _ = chi2_contingency(observed)

# Print the test statistic and p-value
print("Chi-square statistic:", chi2)
print("P-value:", p)
```

p value < significance. We'll reject the null hypothesis. So there's a strong correlation between person's age and loan defaults

### annual income vs loan status

```python
import matplotlib.pyplot as plt

# Extract relevant data for annual income and loan status
incomes = [row['person_income'] for row in data]
loan_statuses = [int(row['loan_status']) for row in data]

# Define income groups
income_groups = ['< $25k', '$25k - $50k', '$50k - $75k', '$75k - $100k', '>$100k']
income_ranges = [(0, 25000), (25000, 50000), (50000, 75000), (75000, 100000), (100000, float('inf'))]

# Group incomes into income ranges and corresponding loan statuses
income_loan_statuses = {}
for income, loan_status in zip(incomes, loan_statuses):
    for i, (start, end) in enumerate(income_ranges):
        if start <= income < end:
            income_group = income_groups[i]
            if income_group in income_loan_statuses:
                income_loan_statuses[income_group]['total'] += 1
                income_loan_statuses[income_group]['defaults'] += loan_status
            else:
                income_loan_statuses[income_group] = {'total': 1, 'defaults': loan_status}
            break

# Calculate proportion of defaulted loans for each income group
income_default_proportion = {income_group: income_info['defaults'] / income_info['total'] for income_group, income_info in income_loan_statuses.items() if income_info['total'] != 0}

# Sort income groups based on proportion of defaults
sorted_income_groups = sorted(income_default_proportion.keys(), key=lambda x: income_default_proportion[x])
sorted_proportions = [income_default_proportion[income_group] for income_group in sorted_income_groups]
sorted_non_default_proportions = [1 - proportion for proportion in sorted_proportions]

# Create stacked bar chart
plt.figure(figsize=(10, 6))
plt.bar(sorted_income_groups, sorted_non_default_proportions, color='lightgreen', label='Non-Defaulted')
plt.bar(sorted_income_groups, sorted_proportions, bottom=sorted_non_default_proportions, color='lightcoral', label='Defaulted')
plt.xlabel('Annual Income Group')
plt.ylabel('Loan Status')
plt.title('Stacked Bar Chart: Annual Income vs Loan Status')
plt.legend()
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

Null Hypothesis (H0): There is no association between income groups and loan defaults.

Alternative Hypothesis (H1): There is an association between income groups and loan defaults.

```python
from scipy.stats import chi2_contingency

# Categorize individuals into income groups
# we can define income groups based on specific thresholds or percentiles
# For example, low-income, medium-income, high-income
# For demonstration, let's categorize into two income groups: low-income and high-income
low_income_defaults = []
high_income_defaults = []

# Define the threshold for distinguishing low-income and high-income individuals
income_threshold = 50000  # For example, individuals earning less than $50,000 are considered low-income

# Iterate through the data and categorize individuals based on annual income
for row in data:
    income = int(row['person_income'])
    if income < income_threshold:
        low_income_defaults.append(int(row['loan_status']))
    else:
        high_income_defaults.append(int(row['loan_status']))

# Count the number of defaults and non-defaults for each income group
observed = [
    [sum(low_income_defaults), len(low_income_defaults) - sum(low_income_defaults)],
    [sum(high_income_defaults), len(high_income_defaults) - sum(high_income_defaults)]
]

# Perform chi-square test
chi2, p, _, _ = chi2_contingency(observed)

# Print the test statistic and p-value
print("Chi-square statistic:", chi2)
print("P-value:", p)
```

With a p-value of 0.0 (essentially 0), it indicates strong evidence against the null hypothesis. Therefore, we reject the null hypothesis and conclude that there is a significant association between income groups and loan defaults.

With a p-value of 0.0 (essentially 0), it indicates strong evidence against the null hypothesis. Therefore, we reject the null hypothesis and conclude that there is a significant association between loan grade and loan defaults.

## hypothesis testing (Kruskal-Wallis-Test)

### age vs loan status

Null Hypothesis (H0): There is no difference in default rates across different age groups.

Alternative Hypothesis (H1): There is a difference in default rates across different age groups.

```python
from scipy.stats import kruskal
from collections import defaultdict

# Assume data is already loaded into `data`

# Define age categories (example ranges)
def categorize_age(age):
    if age < 30:
        return 'Under 30'
    elif age < 50:
        return '30-49'
    else:
        return '50 and over'

# Prepare the data for Kruskal-Wallis test
def prepare_data_for_analysis(data):
    age_group_defaults = defaultdict(list)
    for row in data:
        age_group = categorize_age(int(row['person_age']))
        # Append 1 for default and 0 for non-default
        age_group_defaults[age_group].append(int(row['loan_status']))
    return age_group_defaults

# Perform Kruskal-Wallis test
def perform_kruskal_test(groups):
    # Prepare lists for each age category
    under_30_defaults = groups['Under 30']
    from_30_to_49_defaults = groups['30-49']
    over_50_defaults = groups['50 and over']

    # Conduct the test only if all groups have data
    if len(under_30_defaults) > 0 and len(from_30_to_49_defaults) > 0 and len(over_50_defaults) > 0:
        stat, p_value = kruskal(under_30_defaults, from_30_to_49_defaults, over_50_defaults)
        print("Kruskal-Wallis H statistic:", stat)
        print("P-value:", p_value)

        # Determine if the result is significant
        if p_value < 0.05:
            print("Reject the null hypothesis - there is a significant difference in loan defaults across age groups.")
        else:
            print("Fail to reject the null hypothesis - no significant difference in loan defaults across age groups.")
    else:
        print("Insufficient data in one or more groups to perform the Kruskal-Wallis test.")

# Prepare data
grouped_data = prepare_data_for_analysis(data)

# Perform the test
perform_kruskal_test(grouped_data)
```
